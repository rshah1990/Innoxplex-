```python
import pandas as pd
```

```python
meta_detail = pd.read_csv('titel_description.csv')
```

```python
meta_detail.head()
```

```python
meta_detail['Domain'] = meta_detail['Input URL'].apply(lambda x : x.split("://")[1])
```

```python
unique_domain_df = pd.read_csv('unique_domain_df.csv')
```

```python
unique_domain_df.head()
```

```python
final_df = unique_domain_df.merge(meta_detail,how="outer")
```

```python
final_df.head()
```

```python
# final_df.to_csv(r'C:\Users\rohishah\My pratice ML\analytics vidhya Hackathon\exported\unique.csv',index=False)
```

```python
remaning_df = final_df[final_df['Title'].isnull() & final_df['Meta Description'].isnull() & final_df['Meta Keywords'].isnull()]
```

```python
import time
from bs4 import BeautifulSoup
import requests
```

```python
titl_content = []
start_time = time.clock()
for counter in range(len(remaning_df)):
    title = []
    if counter%30 == 0:
        print("number of iteration{}".format(counter))
        print("time taken {}".format(time.clock() -start_time ))
    row = remaning_df.iloc[counter]
    url = "http"+"://"+row.Domain+"/"
    try:
        r = requests.get(url,timeout=2)
    except:
        print("connection error found http")
        url = "https"+"://"+row.Domain+"/"
        try:
            r = requests.get(url,timeout=2)
        except:
            print("no connection found in https")
            

    soup = BeautifulSoup(r.text, "html.parser")
    metadata = soup.find_all("meta")
    for link in metadata:
        if link.get("name") == "title":
            title.append(link.get("content"))
    titl_content.append(title)
```
